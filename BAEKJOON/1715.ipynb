# N개의 숫자 카드 묶음을 2 묶음씩 합칠 때, 최소값

## 힌트: 그리디 서치, 힙

### 풀이
1. 두 개의 ***가장 작은 묶음***을 꺼낸다.

2. 합치고, 새로운 묶음을 다시 힙에 추가

3. 합칠 때 만큼의 횟수를 sum

```python
import heapq

n = int(input())
dummies = []
sum = 0

# n개의 카드 더미 받아오기
for _ in range(n):
    dummies.append(int(input()))


heapq.heapify(dummies) # 힙 변환
while len(dummies) > 1:
    # 가장 작은 카드 더미 2개 뽑아오기
    d1 = heapq.heappop(dummies)
    d2 = heapq.heappop(dummies)

    # 합치고, 합칠 때 만큼의 횟수를 sum
    cur = d1 + d2

    # 새로운 묶음을 다시 힙에 추가
    heapq.heappush(dummies, cur)

    # 합칠 때 비교 횟수를 sum
    sum += cur


print(sum)
```
