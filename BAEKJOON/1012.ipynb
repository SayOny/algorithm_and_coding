```python
def bfs(x, y, visited, m, n, cabbage):
    # 시작 지점 설정
    queue = [(x, y)]
    visited[x][y] = True

    while queue:
        x, y = queue.pop(0) # 시작

        # 어쨌든 상하좌우에 배추가 있으면 그 배추에는 지렁이를 둘 필요가 없다.
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = x + dx, y + dy # 위치 지정

            # 방문하지 않은 배추에 대해, 배추가 있으면 ? ㄱㄱ
            '''
            조건식에서 &는 비트 연산자로, 논리 연산자를 사용해야 하는 곳에 &를 사용하면 의도치 않은 동작이 발생할 수 있어 and를 사용해야 한다. & 연산자는 비트 연산자로, boolean 값의 논리 연산이 아닌 비트 연산을 수행하기 때문에, 배열의 인덱스 범위 검사를 제대로 수행하지 못할 수 있다.
            '''
            if (0 <= nx < m) and (0 <= ny < n) and (not visited[nx][ny]) and ((nx, ny) in cabbage):
                visited[nx][ny] = True
                queue.append((nx, ny))

t = int(input())

tr = []

for _ in range(t):
    # 배추 밭의 가로 길이, 세로 길이, 심어진 배추의 개수 받아오기
    m, n, k = map(int, input().split(' '))

    # k개의 배추 위치 받아오기
    cabbage = set()
    for _ in range(k):
        x, y = map(int, input().split(' '))
        cabbage.add((x, y))

    # 매츄릭수 만들기
    visited = [[False] * n for _ in range(m)]
    num_worm = 0

    for x, y in cabbage:
        if not visited[x][y]:
            bfs(x, y, visited, m, n, cabbage)
            num_worm += 1

    tr.append(num_worm)

for i in tr:
    print(i)

```
