```python
grades = []
for _ in range(20):
    grades.append(input().split()[1:])

grade_dict = {'A+': 4.5, 'A0': 4.0, 'B+': 3.5,
                    'B0': 3.0, 'C+': 2.5, 'C0': 2.0,
                    'D+': 1.5, 'D0': 1.0, 'F': 0.0}

credit_sum = 0
weighted_sum = 0

for g in grades:
    credit = float(g[0])
    grade = g[1]

    if grade != 'P':
        credit_sum += credit
        weighted_sum += credit * grade_dict[grade]

if credit_sum == 0:
    print(0.0)
else:
    gpa = weighted_sum / credit_sum
    print(f"{gpa:.6f}")
```
