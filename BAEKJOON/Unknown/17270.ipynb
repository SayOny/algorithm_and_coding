1. 지헌, 성하가 있는 노드는 약속 장소로 지정 X
2. 지헌 + 성하 = 최소
3. 지헌 거리 <= 성하 거리
4. [] 지헌에 최단
5. 가장 작은 번호

```python
import heapq
from collections import defaultdict

def dijkstra_all_paths(graph, start, end):
    n = len(graph)

    distances = [float('inf')] * n
    distances[start] = 0

    # 경로 추적을 위한 구조
    paths = defaultdict(list)
    paths[start].append([start])

    # 우선순위 큐를 사용하여 최소 거리 노드를 선택
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # 현재 노드까지의 거리보다 더 짧은 경로가 이미 발견된 경우
        if current_distance > distances[current_node]:
            continue

        # 인접 노드에 대한 거리 계산
        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight

            # 더 짧은 경로가 발견된 경우
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

                # 새로운 최단 경로 추가
                paths[neighbor] = [path + [neighbor] for path in paths[current_node]]

            # 현재 경로와 동일한 길이의 최단 경로가 발견된 경우
            elif distance == distances[neighbor]:
                paths[neighbor].extend(path + [neighbor] for path in paths[current_node])

    return paths[end]

def find_sublist_with_value(lst, value):
    return [sublist for sublist in lst if sublist and sublist[0] == value]


# 시작
v, m = map(int, input().split())

# 그래프 그립시다
graph = [[] for _ in range(v+1)]
for _ in range(1, m+1):
    a, b, c = map(int, input().split())
    graph[a].append([b, c])
    graph[b].append([a, c])

j, s = map(int, input().split())

# v, m = 8, 10

# graph = [[],
#  [[2, 2], [3, 1]],
#  [[1, 2], [6, 3], [7, 2]],
#  [[1, 1], [7, 2]],
#  [[7, 5]],
#  [[6, 2], [7, 2], [8, 2]],
#  [[2, 3], [5, 2]],
#  [[2, 2], [3, 2], [4, 5], [5, 2], [8, 2]],
#  [[7, 2], [5, 2]]]

# j, s = 3, 6


# 지헌이에게서 성하로 가는 최단 경로들을 받아옵니다.
shortest_paths = dijkstra_all_paths(graph, j, s)



distances = {}

for shortest_path in shortest_paths:
    # 지헌이랑 성하가 있는 노드 제외
    if len(shortest_path) <= 2:
        continue
    # shortest_path_without = shortest_path[1:-1]
    [3, 1, 2, 6]

    for n in range(1, len(shortest_path)-1):
        distance_j = 0
        distance_s = 0

        # j와 의 거리 재기
        for i in range(1, n+1):
            distance_j += find_sublist_with_value(graph[shortest_path[i]], shortest_path[i-1])[0][1]

        # s와의 거리 재기
        for i in range(n, len(shortest_path)-1):
            distance_s += find_sublist_with_value(graph[shortest_path[i]], shortest_path[i+1])[0][1]

        if (distance_j <= distance_s):
            if (shortest_path[n] in distances) and (distances[shortest_path[n]] < distance_j):
                pass
            else:
                distances[shortest_path[n]] = distance_j


if not distances:
    print(-1)
elif (len(distances) > 1):
    shortest = {k: v for k, v in distances.items() if v == min(distances.values())}
    if (len(shortest) > 1):
        print(min(shortest.keys()))
    else:
        print(list(shortest.keys())[0])
else:
    print(list(distances.keys())[0])

```
