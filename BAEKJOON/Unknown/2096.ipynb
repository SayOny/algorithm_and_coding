```python
n = int(input())
grid = list(map(int, input().split()))

max_values = grid
min_values = grid

for _ in range(1, n):
    arr = list(map(int, input().split()))

    max_values = [
        arr[0] + max(max_values[0], max_values[1]),
        arr[1] + max(max_values[0], max_values[1], max_values[2]),
        arr[2] + max(max_values[1], max_values[2])
    ]

    min_values = [
        arr[0] + min(min_values[0], min_values[1]),
        arr[1] + min(min_values[0], min_values[1], min_values[2]),
        arr[2] + min(min_values[1], min_values[2])
    ]

print(max(max_values), min(min_values))
```

```python
n = int(input())
grid = [list(map(int, input().split())) for _ in range(n)]

max_values = [grid[0][:]]
min_values = [grid[0][:]]

for i in range(1, n):
    tmp_max = [0] * 3
    tmp_min = [0] * 3

    # 위 줄에서 0, 1번에서 왔을 것
    tmp_max[0] = max(max_values[i-1][0], max_values[i-1][1]) + grid[i][0]
    # 위 줄에서 0, 1, 2번에서 왔을 것
    tmp_max[1] = max(max_values[i-1][0], max_values[i-1][1], max_values[i-1][2]) + grid[i][1]
    # 위 줄에서 1, 2번에서 왔을 것
    tmp_max[2] = max(max_values[i-1][1], max_values[i-1][2]) + grid[i][2]

    tmp_min[0] = min(min_values[i-1][0], min_values[i-1][1]) + grid[i][0]
    tmp_min[1] = min(min_values[i-1][0], min_values[i-1][1], min_values[i-1][2]) + grid[i][1]
    tmp_min[2] = min(min_values[i-1][1], min_values[i-1][2]) + grid[i][2]

    max_values.append(tmp_max)
    min_values.append(tmp_min)

print(max(max_values[n-1]), min(min_values[n-1]))
```

```python
n = int(input())
grid = list(map(int, input().split()))

max_values = [[0] * 3]
min_values = [[0] * 3]

max_values[0] = [grid[0], grid[1], grid[2]]
min_values[0] = [grid[0], grid[1], grid[2]]

for i in range(1, n):
    tmp_max = [0] * 3
    tmp_min = [0] * 3

    if n == 1:
        tmp_max[0] = max(max_values[i-1][0], max_values[i-1][1]) + grid[0]
        tmp_max[1] = max(max_values[i-1][0], max_values[i-1][1], max_values[i-1][2]) + grid[1]
        tmp_max[2] = max(max_values[i-1][1], max_values[i-1][2]) + grid[2]

        tmp_min[0] = min(min_values[i-1][0], min_values[i-1][1]) + grid[0]
        tmp_min[1] = min(min_values[i-1][0], min_values[i-1][1], min_values[i-1][2]) + grid[1]
        tmp_min[2] = min(min_values[i-1][1], min_values[i-1][2]) + grid[2]

    else:
        a, b, c = map(int, input().split())

        tmp_max[0] = max(max_values[i-1][0], max_values[i-1][1]) + a
        tmp_max[1] = max(max_values[i-1][0], max_values[i-1][1], max_values[i-1][2]) + b
        tmp_max[2] = max(max_values[i-1][1], max_values[i-1][2]) + c

        tmp_min[0] = min(min_values[i-1][0], min_values[i-1][1]) + a
        tmp_min[1] = min(min_values[i-1][0], min_values[i-1][1], min_values[i-1][2]) + b
        tmp_min[2] = min(min_values[i-1][1], min_values[i-1][2]) + c

    max_values.append(tmp_max)
    min_values.append(tmp_min)

print(max(max_values[n-1]), min(min_values[n-1]))
```
