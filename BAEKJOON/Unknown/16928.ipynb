```python
def bfs(n, m, board):
    queue = [(1, 0)]  # (cur, 주사위 굴린 횟수)

    # 방문 여부
    visited = [False] * 101
    visited[1] = True

    while queue:
        cur, num = queue.pop(0)

        # 딱 100이 되면 나가자 (100이 안되는 경우는 없음)
        if cur == 100:
            return num

        for i in range(1, 7):
            next = cur + i
            if next <= 100 & !(visited[board[next]]):
                visited[board[next]] = True
                queue.append((board[next], num + 1))
    return -1



n, m = map(int, input().split(' '))

ladders = []
snakes = []

for _ in range(n):
    x, y = map(int,input().split(' '))
    ladders.append((x, y))

for _ in range(m):
    u, v = map(int,input().split(' '))
    snakes.append((u, v))

board = list(range(101))

# 사다리와 뱀이 걸렸을 때 이동하는 위치 명시
for x, y in ladders:
    board[x] = y

for u, v in snakes:
    board[u] = v

print(bfs(n, m, board))
```
