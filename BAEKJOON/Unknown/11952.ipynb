```python
import heapq

def bfs(path, zombie_city, n, s):
    '''
        각 노드들이 좀비시티로부터 얼마나 떨어져있는지
        분명 더 좋은 방법이 있을 것 같은데.. ㅎㄱㅎ극
        생각이 안 남
    '''

    dist = [float('inf')] * (n + 1)
    queue = []

    for z in zombie_city:
        queue.append((z, 0))
        dist[z] = 0

    popleft = 0
    while queue:
        try:
            cur, d = queue[popleft]
        except IndexError:
            break

        popleft += 1
        for i, _ in path[cur]:
            if dist[i] > d + 1: # 멀면 기강 잡기, 좀비로부터 벗어날 수 없다
                dist[i] = d + 1
                queue.append((i, d + 1))

    return dist

def dijkstra(path, start, n):
    '''
        시작 노드로부터 가장 비용이 적은 연결된 도로를 선택해서
        쫙쫙 선택해 나가는 거
    '''
    pq = [(0, start)]
    min = [float('inf')] * (n + 1)
    min[start] = 0

    while pq:
        cur, u = heapq.heappop(pq)
        if cur <= min[u]:
            for i, cost in path[u]: # u랑 연결되어 있는 애들이랑 더해서
                new = cur + cost
                if new < min[i]: # i번째 최소값이면 업데이트
                    min[i] = new
                    heapq.heappush(pq, (new, i))
        else:
            continue

    return min[n]




n, m, k, s = map(int, input().split())
p, q = map(int, input().split())

zombie_city = []
for _ in range(k):
    zombie_city.append(int(input()))

path = [[] for _ in range(n+1)]
for _ in range(m):
    u, v = map(int, input().split())
    path[u].append((v, 0))
    path[v].append((u, 0))


# 좀비시티와 인접한 s개의 도시 찾고, 모든 도시들의 비용을 세팅 해주자
dist = bfs(path, zombie_city, n, s)
# 아마 여기에서 문제인 듯...............
for i in range(1, n+1):
    for j in range(len(path[i])):
        v, _ = path[i][j]
        path[i][j] = (v, (q if dist[i] <= s else p))

# 최소 비용의 길을 찾자
min_cost = dijkstra(path, 1, n)

print(min_cost)

```
