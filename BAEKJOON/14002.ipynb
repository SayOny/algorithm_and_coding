```python
n = int(input().strip()) # 수열의 길이 입력받기
a = list(map(int, input().strip().split())) # 수열 공백으로 띄어서 입력받기

# 1. 초기화
# 길이가 0일 경우를 대비한 나머지를 inf로 해서 모든 원소가 작을 수 있도록 초기화
l = [-float('inf')] + [float('inf')] * n
# 각 시점에서의 값의 인덱스 넣기~
indexes = [0] * n

def binary_search(l, num, low, high):
    while low < high:
        mid = (low + high) // 2
        if l[mid] < num:
            low = mid + 1
        else:
            high = mid
    return low


# 2. a[i] 값의 위치 찾기
length = 0
for i in range(n):
    num = a[i]
    idx = binary_search(l, num, 1, length + 1)
    l[idx] = num
    indexes[i] = idx
    if idx > length:
        length = idx

# 3. 부분 수열 재구성
new = []
for i in range(n - 1, -1, -1):
    if indexes[i] == length:
        new.append(a[i])
        length -= 1
        if length == 0:
            break

new.reverse()


print(len(new))
print(" ".join(map(str, new)))
```
