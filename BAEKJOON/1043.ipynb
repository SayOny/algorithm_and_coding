```python
n, m = map(int, input().split())

knows = list(map(int, input().split()))[1:]

had_came = set() # 중복을 허용하지 않는 자료구조를 이용해 (거짓말 하면 안되는)파티에 오는 사람들을 업데이트

lie = 0 # 거짓말 해도 되는 파티 수

for _ in range(m):
    nums = list(map(int, input().split()))[1:] # 각 파티에 오는 사람의 번호
    can_lie = True # 거짓말 할 수 있는 상태인지

    for n in nums:
        if n in knows:
            can_lie = False
            had_came.update(nums)

        elif n in had_came:
            can_lie = False
            had_came.update(nums)

    if can_lie:
        lie += 1


print(lie)
```

```python
n, m = map(int, input().split())

knows = set(list(map(int, input().split()))[1:])

had_came = set()
lie = 0

for _ in range(m):
    nums = set(list(map(int, input().split()))[1:])

    if knows & nums:  # 교집합, 겹치는 요소가 하나라도 있으면 True
        had_came.update(nums)
        continue
    elif had_came & nums:
        had_came.update(nums)
        continue

    lie += 1

print(lie)
```

# 파티의 모든 순서를 고려해야 하는구나

```python
n, m = map(int, input().split())

knows = set(list(map(int, input().split()))[1:]) # 0인 경우 set 변환 에러(set는 인덱싱 X) -> 리스트로 슬라이싱 후 세트로 변환

lie = 0

parties = []
for _ in range(m):
    party = set(list(map(int, input().split()))[1:])
    parties.append(party)

# 모든 파티 순서를 고려
for _ in range(m):
    for party in parties:
        if party & knows:
            knows.update(party) # update는 원래 세트를 변경하는 반면, union은 새로운 세트를 생성

for party in parties:
    if party & knows:
        continue
    lie += 1

print(lie)
```
