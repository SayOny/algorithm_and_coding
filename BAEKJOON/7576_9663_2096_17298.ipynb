# 7576 토마토

```python
from collections import deque

def bfs():
    queue = deque()

    # 처음에 익은 토마토들을 큐에 넣음
    for i in range(N):
        for j in range(M):
            if grid[i][j] == 1:
                queue.append((i, j))

    # 익은 토마토들을 기준으로 탐색
    while queue:
        x, y = queue.popleft()

        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]: # 위아래양옆으로
            nx, ny = x + dx, y + dy
            # 인접한 칸이 익지 않은 토마토면 익히고 날짜 추가
            if 0 <= nx < N and 0 <= ny < M and grid[nx][ny] == 0:
                grid[nx][ny] = grid[x][y] + 1  # 익은 날짜는 현재 날짜 + 1
                queue.append((nx, ny))

def tomatoes():
    # 모든 칸을 순회하며 익지 않은 토마토가 있는지 확인
    max_day = 0
    for i in range(N):
        for j in range(M):
            if grid[i][j] == 0:
                return -1  # 익지 않은 토마토가 있는 경우
            max_day = max(max_day, grid[i][j])  # 가장 늦게 익은 날짜 갱신
    return max_day - 1

M, N = map(int, input().split())
grid = [list(map(int, input().split())) for _ in range(N)]

# 익히기
bfs()

# 결과 출력
result = tomatoes()
print(result)
```

# 2096 내려가기

```python
n = int(input())
grid = [list(map(int, input().split())) for _ in range(n)]

max_values = [grid[0][:]]
min_values = [grid[0][:]]

for i in range(1, n):
    tmp_max = [0] * 3
    tmp_min = [0] * 3

    # 위 줄에서 0, 1번에서 왔을 것
    tmp_max[0] = max(max_values[i-1][0], max_values[i-1][1]) + grid[i][0]
    # 위 줄에서 0, 1, 2번에서 왔을 것
    tmp_max[1] = max(max_values[i-1][0], max_values[i-1][1], max_values[i-1][2]) + grid[i][1]
    # 위 줄에서 1, 2번에서 왔을 것
    tmp_max[2] = max(max_values[i-1][1], max_values[i-1][2]) + grid[i][2]

    tmp_min[0] = min(min_values[i-1][0], min_values[i-1][1]) + grid[i][0]
    tmp_min[1] = min(min_values[i-1][0], min_values[i-1][1], min_values[i-1][2]) + grid[i][1]
    tmp_min[2] = min(min_values[i-1][1], min_values[i-1][2]) + grid[i][2]

    max_values.append(tmp_max)
    min_values.append(tmp_min)

print(max(max_values[n-1]), min(min_values[n-1]))
```

# 17298 오큰수

```python
def find_greater(A):
    N = len(A)
    result = [-1] * N
    stack = []  # 인덱스 저장

    for i in range(N):
        # 현재 수가 스택 뚜껑 수보다 크면
        while stack and A[stack[-1]] < A[i]:
            idx = stack.pop()
            result[idx] = A[i]

        stack.append(i)

    return result

N = int(input())
A = list(map(int, input().split()))

result = find_greater(A)

print(" ".join(map(str, result)))
```
