# 1446 지름길

### 문제
    매일 아침, 세준이는 학교에 가기 위해서 차를 타고 D킬로미터 길이의 고속도로를 지난다. 이 고속도로는 심각하게 커브가 많아서 정말 운전하기도 힘들다. 어느 날, 세준이는 이 고속도로에 지름길이 존재한다는 것을 알게 되었다. 모든 지름길은 일방통행이고, 고속도로를 역주행할 수는 없다.

    세준이가 운전해야 하는 거리의 최솟값을 출력하시오.

### 입력
    첫째 줄에 지름길의 개수 N과 고속도로의 길이 D가 주어진다. N은 12 이하인 양의 정수이고, D는 10,000보다 작거나 같은 자연수이다. 다음 N개의 줄에 지름길의 시작 위치, 도착 위치, 지름길의 길이가 주어진다. 모든 위치와 길이는 10,000보다 작거나 같은 음이 아닌 정수이다. 지름길의 시작 위치는 도착 위치보다 작다.

### 출력
    세준이가 운전해야하는 거리의 최솟값을 출력하시오.

```python
n, d = map(int, input().split())

# 다익스트라 알고리즘을 위한 그래프 그리기, 딕셔너리로
graph = {i: [] for i in range(d + 1)}  
for i in range(n):
    start, end, dist = map(int, input().split())
    graph[start].append((end, dist))
```

그래프를 어떻게 그릴지 고민했는데, 그냥 거리 1 단위마다 노드를 설정하는 것이 편하겠다.

```python
graph
```

```python
def find_shortest():
    # 최소 힙 씁니다.
    pq = [(0, 0)]  # (현재 까지의 distance, 현재 노드)
    distances = [float('inf')] * (d + 1)  # 각 지점에 대한 최단 거리를 위해 국룰 무한대로 초기화 해줍니다.
    distances[0] = 0
    
    while pq:
        pq.sort()  # 가장 작은 거리의 노드를 찾기 위해 정렬합니다
        cur_dist, cur_node = pq.pop(0) # 거리가 가장 작은 노드 꺼내와요
        
        # 이 노드가 목적지라면 리턴
        if cur_node == d:
            return 
        
        # 이미 더 작은 거리(가까운 경로들로 온 경우)가 있는 경우
        if cur_dist > distances[cur_node]:
            continue
        
        # 걍 고속도로 길로 가는 경우 
        if cur_node + 1 <= d and distances[cur_node] + 1 < distances[cur_node + 1]:
            distances[cur_node + 1] = distances[cur_node] + 1 # 말 그대로 1을 더하는 것
            pq.append((distances[cur_node + 1], cur_node + 1)) # 힙에 추가
        
        # 지름길로 가는 경우
        for end, distance in graph[cur_node]:
            if end <= d and cur_dist + distance < distances[end]:
                distances[end] = cur_dist + distance 
                pq.append((distances[end], end)) # 힙에 추가
    
    return distances[d] # 결국 위에서 계속 업데이트 된 distances를 이용

result = find_shortest()
print(result)
```

```python
n, d = map(int, input().split())

# 다익스트라 알고리즘을 위한 그래프 그리기, 딕셔너리로
graph = {i: [] for i in range(d + 1)}  

for i in range(n):
    start, end, dist = map(int, input().split())
    if end <= d:  # end가 d보다 크면 그건 버림
        graph[start].append((end, dist))

def find_shortest():
    # 최소 힙 씁니다.
    pq = [(0, 0)]  # (현재 까지의 distance, 현재 노드)
    distances = [float('inf')] * (d + 1)  # 각 지점에 대한 최단 거리를 위해 국룰 무한대로 초기화 해줍니다.
    distances[0] = 0
    
    while pq:
        pq.sort()  # 가장 작은 거리의 노드를 찾기 위해 정렬합니다
        cur_dist, cur_node = pq.pop(0)  # 거리가 가장 작은 노드 꺼내와요
        
        # 이 노드가 목적지라면 리턴
        if cur_node == d:
            return 
        
        # 이미 더 작은 거리가 있는 경우
        if cur_dist > distances[cur_node]:
            continue
        
        # 걍 고속도로 길로 가는 경우 
        if cur_node + 1 <= d and distances[cur_node] + 1 < distances[cur_node + 1]:
            distances[cur_node + 1] = distances[cur_node] + 1 # 말 그대로 1을 더하는 것
            pq.append((distances[cur_node + 1], cur_node + 1)) # 힙에 추가
        
        # 지름길로 가는 경우
        for end, distance in graph[cur_node]:
            if end <= d and cur_dist + distance < distances[end]:
                distances[end] = cur_dist + distance 
                pq.append((distances[end], end)) # 힙에 추가
    
    return distances[d] # 결국 위에서 계속 업데이트 된 distances를 이용

result = find_shortest()
print(result)
```
