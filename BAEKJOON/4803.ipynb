트리는
1. 노드가 n개면, 브랜치가 n-1이다.
2. 싸이클이 없어야 한다.
3. 노드가 1개면, 브랜치가 0인 트리로 간주.

```python
def dfs(graph, i):
    stack = [i]
    visited_new = set() # 이번 dfs에서 방문하는 노드
    numOfBranches = 0

    while stack:
        cur = stack.pop()
        if cur in visited_new:
            continue
        visited_new.add(cur)

        for n in graph[cur]:
            if n not in visited_new:
                stack.append(n)
            numOfBranches += 1

    numOfBranches //= 2

    return visited_new, numOfBranches




numOfcases = 1 # 테스트 케이스

while(True):
    # 노드와 브랜치 개수 가져오기
    n, m = map(int, input().split())

    # 0 0 들어왔으면 종료
    if(n == 0 and m == 0):
        break

    # 브랜치 정보 가져오기
    graph = [[] for _ in range(n+1)] #인접리스트
    for _ in range(m):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)


    visited = set() # 방문한 노드 넣을 곳
    numOfTrees = 0 # 트리 개수
    for i in range(1, n + 1):
        if i not in visited:
            visited_new, numOfBranches = dfs(graph, i)
            visited.update(visited_new) # visited 업데이트
            # 새로 탐색한 노드들에 대해서, 브랜치 수가 노드보다 하나 적으면 트리가 맞다고 판단.
            if(len(visited_new) - 1 == numOfBranches):
                numOfTrees += 1

    if numOfTrees == 0:
        print(f"Case {numOfcases}: No trees.")
    elif numOfTrees == 1:
        print(f"Case {numOfcases}: There is one tree.")
    else:
        print(f"Case {numOfcases}: A forest of {numOfTrees} trees.")

    numOfcases += 1

```
