# 17298 오큰수

```python
def find_greater(A):
    N = len(A)
    result = [-1] * N
    stack = []  # 인덱스 저장

    for i in range(N):
        # 현재 수가 스택 뚜껑 수보다 크면
        while stack and A[stack[-1]] < A[i]:
            idx = stack.pop()
            result[idx] = A[i]

        stack.append(i)

    return result

N = int(input())
A = list(map(int, input().split()))

result = find_greater(A)

print(" ".join(map(str, result)))
```
