# 17298 오큰수

힌트가 스택이다.
1. n, a 받아오기 아니 맨날 n을 왜 받아오는거임? -> 다른 언어에서는 길이 세는게 빡셀 수 있으니까 그런가보다.
2. 오큰수 리스트(NGE): 기본 값이 -1인 길이 n의 리스트로, 각 인덱스에 대한 오큰수를 저장
3. 스택을 사용하여 큰 수 못만나면 짱박혀서 다음 수 기다림
4. 오큰수 없으면 기본 값 -1 유지, 그리고 마지막 인덱스의 오큰수는 무조건 -1

```python

n = int(input())
a = list(map(int, input().split()))

NGE = [-1] * n

stack = []  # 인덱스가 저장됨
for i in range(n):
    # 여태까지 앞의 인덱스들 중 이후 인덱스의 값들에서 큰 수를 모두 못 찾은 스택 값들에 대해 비교
    while stack:
        if a[stack[-1]] < a[i]:
            index = stack.pop()
            NGE[index] = a[i]
    stack.append(i) # i번째 값을 스택에 넣고 다음 수들 부터 비교

print(" ".join(map(str, NGE)))
```
